```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data
```

## Player Count

* Display the total number of players


```python
unique_players = purchase_data['SN'].nunique()
players_df = pd.DataFrame({
    'Players': [unique_players]
})
players_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_items = purchase_data['Item Name'].nunique()
ave_price = purchase_data['Price'].mean()
total_sales = purchase_data['Price'].sum()
salesummary_df = pd.DataFrame({
    'Unique Items': [unique_items],
    'Average Sale Price': [ave_price],
    'Total Sales': [total_sales]
})
salesummary_df.style.format({"Average Sale Price": "${:20,.2f}",
                            "Total Sales": "${:20,.0f}"})
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
unique_players_df = purchase_data.filter(['SN', 'Gender'])
unique_players_gender = unique_players_df.drop_duplicates(subset='SN', keep='last')
male_data = unique_players_gender[unique_players_gender["Gender"] == "Male"].count()["Gender"]
female_data = unique_players_gender[unique_players_gender["Gender"] == "Female"].count()["Gender"]
nondisclose_data = unique_players - male_data - female_data
per_male = male_data / unique_players * 100
per_female = female_data / unique_players * 100
per_non = nondisclose_data / unique_players * 100
raw_df_data = {
    'Number of Players': [male_data, female_data, nondisclose_data],
    'Percentage of Players': [per_male, per_female, per_non]
}
gender_df = pd.DataFrame(raw_df_data, index = ['Male', 'Female', 'Other / Non-Disclosed'])
gender_df.style.format({"Percentage of Players": "{:20,.2f}%"})
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
gender_purchase_df = purchase_data.filter(['SN', 'Gender', 'Price'])
male_purchase_count = gender_purchase_df[gender_purchase_df["Gender"] == "Male"].count()["SN"]
male_purchase_sum = gender_purchase_df[gender_purchase_df["Gender"] == "Male"].sum()["Price"]
male_purchase_avg = male_purchase_sum / male_purchase_count
female_purchase_count = gender_purchase_df[gender_purchase_df["Gender"] == "Female"].count()["SN"]
female_purchase_sum = gender_purchase_df[gender_purchase_df["Gender"] == "Female"].sum()["Price"]
female_purchase_avg = female_purchase_sum / female_purchase_count
other_purchase_count = gender_purchase_df[gender_purchase_df["Gender"] == "Other / Non-Disclosed"].count()["SN"]
other_purchase_sum = gender_purchase_df[gender_purchase_df["Gender"] == "Other / Non-Disclosed"].sum()["Price"]
other_purchase_avg = other_purchase_sum / other_purchase_count
df_data = {
    'Purchase Count': [male_purchase_count, female_purchase_count, other_purchase_count],
    'Average Purchase': [male_purchase_avg, female_purchase_avg, other_purchase_avg],
    'Total Purchase': [male_purchase_sum, female_purchase_sum, other_purchase_sum]
}
gender_analysis_df = pd.DataFrame(df_data, index = ['Male', 'Female', 'Other / Non Disclosed'])
gender_analysis_df.style.format({"Average Purchase":"${:20,.2f}", "Total Purchase":"${:20,.2f}"})
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
age_pur_df = purchase_data.filter(['SN', 'Age', 'Price', 'Purchase ID'])
age_pur_df["Age Group"] = pd.cut(age_pur_df["Age"], age_bins, labels=age_labels)

age_pur_group = age_pur_df.groupby(['Age Group']).agg({'Price':'sum', 'Purchase ID':'count'})
age_pur_group['Average Purchase'] = age_pur_group['Price']/age_pur_group['Purchase ID']
age_rename = age_pur_group.rename(columns={"Price": "Total Spend", "Purchase ID": "Purchase Count"})

age_rename.style.format({"Total Spend":"${:20,.0f}", "Average Purchase":"${:20,.2f}"})
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
top_buyer_df = purchase_data.filter(['SN', 'Price'])
top_buyer_df["Price"] = pd.to_numeric(top_buyer_df["Price"], downcast="float")
top_buyercon_df = top_buyer_df.groupby(['SN']).agg({'Price':'sum', 'SN':'count'})

top_buyer_format = top_buyercon_df.rename(columns={"Price": "Total Spend", "SN": "Purchase Count"})
top_buyer_format['Average Purchase'] = top_buyer_format['Total Spend']/top_buyer_format['Purchase Count']
top_buyer_sort = top_buyer_format.sort_values(by='Total Spend', ascending=False)
top_buyer_final = top_buyer_sort.head().style.format({"Total Spend":"${:20,.2f}", "Average Purchase":"${:20,.2f}"})
top_buyer_final
```
